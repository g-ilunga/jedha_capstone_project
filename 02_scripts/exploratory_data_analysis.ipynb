```python
#=====================================
# Libraries
#=====================================
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#========================
# Loading dataset
#========================
dataset = pd.read_excel("C:/Users/ilung/Documents/Jedha_bootcamp/capstone_project/data/clean/france/year.xlsx")
```

### Computing some statistics

```python
pollutants_list = ["no2", "o3", "pm2p5", "pm10", "so2"]

# For each pollutant, finding the minimum concentration for each pollutant and the corresponding city
for pollutant in pollutants_list:
    min_concentration = dataset.loc[dataset[pollutant].idxmin(), ["name", pollutant]]
    print (f"The lowest {pollutant} concentration is {min_concentration[pollutant]} ug/m3 and is recorded at {min_concentration['name']}")

print("=" * 100) 

# For each pollutant, finding the maximum concentration for each pollutant and the corresponding city
for pollutant in pollutants_list:
    min_concentration = dataset.loc[dataset[pollutant].idxmax(), ["name", pollutant]]
    print (f"The highest {pollutant} concentration is {min_concentration[pollutant]} ug/m3 and is recorded at {min_concentration['name']}")
```

### Geospatial distribution of pollutants

```python
pollutants_list = ["no2", "o3", "pm2p5", "pm10", "so2"]

for pollutant in pollutants_list:
    plt.figure(figsize=(12, 6))
    sc = plt.scatter(dataset["longitude"], dataset["latitude"], c=dataset[pollutant], cmap="Reds", alpha=0.7, edgecolors="black")

    plt.colorbar(sc, label=f"{pollutant.upper()} Levels (Âµg/mÂ³)")
    plt.title(f"Geospatial Distribution of {pollutant.upper()} Levels", fontsize=14)
    plt.xlabel("Longitude", fontsize=12)
    plt.ylabel("Latitude", fontsize=12)
    plt.grid(True, linestyle="--", alpha=0.5)
    plt.show()
```

### Correlation matrix between pollutants

```python
pollutant_columns = ["no2", "o3", "pm2p5", "pm10", "so2"]

# Computing correlation matrix
corr_matrix = dataset[pollutant_columns].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", linewidths=0.5, fmt=".2f")
plt.title("Correlation Heatmap of Air Pollutants", fontsize=14)
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.show()
```
